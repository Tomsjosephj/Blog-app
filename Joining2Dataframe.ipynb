```python
#df1 id name 4

import pandas as pd
lis1=[[101,'Toms'],[102,'Akhil'],[103,'Joseph'],[104,'Shanal']]
df1=pd.DataFrame(lis1,columns=['Id','Name'])
lis2=[[105,'Johny'],[106,'Paul'],[107,'Joy'],[108,'Vargheese']]
df2=pd.DataFrame(lis2,columns=['Id','Name'])
df1
```

```python
df2
```

```python
#to combine two dataframe
df3=pd.concat([df1,df2],ignore_index=True)
df3
```

```python
#df4 age cities 4

df4=pd.DataFrame({
    'age':[23,45,34,28],
    'city':['Berlin','Hamburg','Newyork','Tokyo']
})
df4
```

```python
df5= pd.concat([df1, df4], axis=1)
df5
```

```python
#dfa rollno name class

dfa=pd.DataFrame({
    'Roll no':[101,102,104,105],
    'Name':['James','Manu','Maneesh','Vinay'],
    'Class':[12,11,10,12]
})
dfa
```

```python
dfb=pd.DataFrame({
    'Roll no':[101,102,103,104],
    'Batch':['A','A','B','C'],
    'Mark':[45,44,43,42]
})
dfb
```

```python

result=pd.merge(dfa,dfb,how='inner')
result
```

```python

result=pd.merge(dfa,dfb,how='left')
result
```

```python
result=pd.merge(dfa,dfb,how='right')
result
```

```python
result=pd.merge(dfa,dfb,how='outer')
result
```
